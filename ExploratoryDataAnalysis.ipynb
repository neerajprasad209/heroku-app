# Performing ANOVA with continuous variables

```python
from definitions import ANOVA
q = []
for i in con:
    q.append(ANOVA(df1,"default",i))
```

```python
anova_df = pd.DataFrame({'Column':con,'pval':q})
anova_df.sort_values(by='pval')
```

```python
imp_con_columns = list(anova_df[anova_df.pval<0.05]['Column'])
imp_con_columns
```

# Chi Square test for Categorical to Target

```python
from definitions import chisquare
r = []
for i in cat:
    r.append(chisquare(df1,'default',i))
```

```python
chi_df = pd.DataFrame({'Columns':cat,'Pval':r})
chi_df.sort_values(by='Pval')
```

```python
imp_cat_columns = list(chi_df[chi_df['Pval']<0.05]['Columns'])
imp_cat_columns.remove('default')
imp_cat_columns
```
